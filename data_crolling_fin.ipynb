### 함수로 만들어 보기

```python
def getPrice(code):
    url = 'https://finance.naver.com/item/main.nhn?code='+code
    soup = bs(urlopen(url), 'html.parser')
    today = soup.find('div', class_='today')
    today = today.find('span', class_='blind')
    price = today.get_text()+'원'
    print(price)
```

```python
getPrice('068270')
```

```python
def getPriceMktcap(code):
    url = 'https://finance.naver.com/item/main.nhn?code='+code
    soup = bs(urlopen(url), 'html.parser')
    today = soup.find('div', class_='today')
    today = today.find('span', class_='blind')

    mktcap = soup.find('table', summary='시가총액 정보')
    mktcap = mktcap.find('em', id='_market_sum').get_text().replace('\n', '').replace('\t', '')
    print(f'--> {code} 기본 정보')
    print('현재가: '+today.get_text()+'원')
    print(f'시가총액: {mktcap} 억원')
```

```python
getPriceMktcap('005930')
```

#### * 내 관심종목 모니터링 하기

```python
for code in ['005930', '005380', '035420']:
    getPriceMktcap(code)
```

### 여러종목을 한번에 긁어와보자

```python
url = 'https://finance.naver.com/sise/sise_market_sum.nhn'
soup = bs(urlopen(url), 'html.parser')
soup = soup.find('table', summary="코스피 시세정보를 선택한 항목에 따라 정보를 제공합니다.")
soup = soup.find('tbody').find_all('tr', onmouseout="mouseOut(this)")

for content in soup[:10]:
    print('* 종목명: '+content.find_all('td')[1].get_text())
    print('   - 현재가: '+content.find_all('td')[2].get_text()+'원')
    print('   - 시가총액: '+content.find_all('td')[6].get_text()+'억원')
    print('   - 외국인 비율: '+content.find_all('td')[8].get_text()+'%')
```
